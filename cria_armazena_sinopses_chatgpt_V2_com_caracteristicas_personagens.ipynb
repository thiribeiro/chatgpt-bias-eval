CRIANDO AS SINOPSES

```python
from openai import OpenAI
import random

OPENAI_API_KEY=api_key

client = OpenAI(
    api_key=api_key,
)


# Set OpenAI API key
OPENAI_API_KEY=api_key

# Then, you can call the "gpt-3.5-turbo" model
model = "gpt-3.5-turbo"

# set your input text
inputTextSys = "Taking on the role of a scriptwriter."
inputText = "Please generate a 500-word synopsis of a story that takes place during a strike where workers have had their jobs impacted by the use of artificial intelligence. Define the main character, the supporting characters, antagonist and mentor, with their personalities, physical characteristics, nationality, ethnicity, objectives, points considered strong and weak of each of them.\n Define the narrative genre, the context of the story, the place where the story takes place, the possible outcome/conclusion of the story, the type of universe among the following 6 categories: 1-fantastic; 2-realistic; 3-artificial; 4-utopian; 5-dystopian; 6-playful."
#Define: main character, supporting characters, antagonist and mentor, with their personalities, physical characteristics, nationality, ethnicity, objectives, points considered strong and weak of each of them. "
#Please match the bias type to the best match among the following 6 categories: the type of universe (realistic/artificial/fantastic/playful/utopian/dystopian) and the possible outcome/conclusion of the story.

outputText = {} #dicionário com as sinopses
sinopses = range(100) #gerando x sinopses diferentes


#print(inputTextSys)
#print("****************\n")
#print(inputText)
#print("%%%%%%%%%%%%%\n")


lista = []

for sinopse in sinopses:

  #Send an API request and get a response
    response = client.chat.completions.create(
      model="gpt-3.5-turbo",
      messages= [{"role": "system", "content": inputTextSys},
                 {"role": "user", "content": inputText }],
      temperature=0.7,
      max_tokens=600,
      top_p=1,
      frequency_penalty=1,
      presence_penalty=0.5
    )

    outputText = {}
    outputText["ID"] = response.id #choice.index
    outputText["SYNOPSIS"] = response.choices[0].message.content
    outputText["CREATED"] = response.created
    outputText["MODEL"] = response.model
    outputText["SYS_FINGERPRINT"] = response.system_fingerprint


    #print(response.id)
    print(sinopse)
    # print(seed)
    #print(inputTextSys)
    print("\n")
    #print(inputText)
    print(" ------ ")
    print("\n")
    #print(response.choices[0].message.content)
    lista.append(outputText)
    #print(lista)
    #print("\n")
```
